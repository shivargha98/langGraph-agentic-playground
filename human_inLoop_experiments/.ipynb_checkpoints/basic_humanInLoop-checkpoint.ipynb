```python
from typing import TypedDict, Annotated, List
from langchain_core.messages import HumanMessage, AIMessage
from langgraph.graph import StateGraph, END, add_messages
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv()
```

```python
llm_model = ChatGroq(model = 'llama-3.1-8b-instant')
class State(TypedDict):
    messages: Annotated[List, add_messages]

GENERATE_POST = "generate_post"
GET_REVIEW_DECISION = "get_review_decision"
POST = "postNode"
COLLECT_FEEDBACK = "collectFeedback_node"

def generate_post(state: State):

    return {

        "messages":[llm_model.invoke(state['messages'])]
    }

def review_current_post(state: State):

    current_post = state["messages"][-1].content
    print("\nðŸ“¢ Current LinkedIn Post:\n")
    print(current_post)
    print("\n")

    decision = input("Do you want to post this (yes/no)?")
    if decision.lower().strip() == "yes":
        return POST
    else:
        return COLLECT_FEEDBACK
```

```python
def postNode(state: State):

    final_post = state['messages'][-1].content
    print("Final Post on linkedin:\n")
    print(final_post)
    print("\nâœ… Post has been approved and is now live on LinkedIn!")

def collectFeedback_node(state: State):

    feedback = input("How can I make this post better?")
    return {

        'messages':[HumanMessage(content = feedback)]
    }
```

```python
## getting all the Nodes ##
graph = StateGraph(State)
graph.add_node(GENERATE_POST,generate_post)
graph.add_node(COLLECT_FEEDBACK,collectFeedback_node)
graph.add_node(POST,postNode)

graph.set_entry_point(GENERATE_POST)

## Setting the edges ##
graph.add_conditional_edges(GENERATE_POST,review_current_post) ## decision maker node ##
#graph.add_conditional_edges(GET_REVIEW_DECISION,review_current_post)
graph.add_edge(POST,END)
graph.add_edge(COLLECT_FEEDBACK,GENERATE_POST)
app = graph.compile()
```

```python
from IPython.display import Image, display
png = app.get_graph().draw_mermaid_png()
display(Image(png))
```

```python
response = app.invoke({
    "messages": [HumanMessage(content="Write me a LinkedIn post on AI Agents taking over content creation")]
})
```
